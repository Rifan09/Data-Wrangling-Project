Setelah melakukan pengecekan pada data loan home sanction terdapat banyak missing value, dan setelah dicek setiap kolom ternyata banyak mengandung informasi penting. jadi saya memutuskan untuk melakukan cleaning data dengan teknik imputation, yaitu mengganti value dari setiap kolom dengan value paling banyak pada kolom tersebut.

```python
loan_df[loan_df.Gender.isna()]
```

```python
loan_df.Gender.value_counts()
```

```python
loan_df['Gender'].fillna(value = "Male", inplace = True)
```

```python
loan_df.info()
```

```python
loan_df.Married.value_counts()
```

```python
loan_df[loan_df.Married.isna()]
```

```python
loan_df['Married'].fillna(value = "Yes", inplace = True)
```

```python
loan_df[loan_df.Dependents.isna()]
```

```python
loan_df.Dependents.value_counts()
```

```python
# Menggunakan mode untuk mengisi nilai yang hilang
mode_value = loan_df['Dependents'].mode()[0]  # Mendapatkan nilai mode
loan_df['Dependents'].fillna(value=mode_value, inplace=True)
```

```python
# Menggunakan mode untuk mengisi nilai yang hilang
mode_value = loan_df['Credit_History'].mode()[0]  # Mendapatkan nilai mode
loan_df['Credit_History'].fillna(value=mode_value, inplace=True)
```

```python
loan_df.info()
```

```python
loan_df.isna().sum()
```

```python
loan_df.Self_Employed.value_counts()
```

```python
loan_df['Self_Employed'].fillna(value = "No", inplace = True)
```

```python
loan_df.Loan_Amount_Term.value_counts()
```

Selain teknik imputation saya juga menggunakan teknik statistika untuk menggati nilai yang hilang seperti mean()

```python
# Menggunakan mean untuk mengisi nilai yang hilang
loan_df['Loan_Amount_Term'].fillna(value=loan_df['Loan_Amount_Term'].mean(), inplace=True)
```

```python
loan_df.LoanAmount.value_counts()
```

```python
# Menggunakan mean untuk mengisi nilai yang hilang
loan_df['LoanAmount'].fillna(value=loan_df['LoanAmount'].mean(), inplace=True)
```
